### Задание 1: Фермер

**Условие:**
На ферме содержатся шесть разных видов животных, и каждый раз, когда фермер заходит в сарай, он видит одно случайное животное. За день фермер заходит в сарай 6 раз. ### Задание 2: Кулинарное соревнование

**Условие:**
В конкурсе участвуют 80 шеф-поваров с уникальными уровнями мастерства. В первом этапе судьи случайным образом распределяют их по парам (в любом состязании двух шефов выигрывает тот, у кого выше уровень мастерства). На втором этапе шефы снова случайно образуют пары для финального раунда (пары могут повториться). Победную награду получают те, кто выиграл в обоих этапах. Каково математическое ожидание числа победителей? Ответ округлить до десятых, например: 33,5.

````python
# Расчет через сумму вероятностей для каждого ранга
n_chefs = 80
expected_winners = sum(((k - 1) / (n_chefs - 1)) ** 2 for k in range(1, n_chefs + 1))

print(f"Ответ: {expected_winners:.1f}".replace('.', ','))
````

Представим, что всех поваров отсортировали по мастерству и дали им ранги от 1 (слабый) до 80 (сильный). Для каждого повара посчитаем вероятность выиграть оба этапа, а потом сложим эти вероятности.

**Логика для одного повара с рангом k:**

- Всего соперников — 79 человек (все, кроме него самого)
- Соперников, которые слабее него — ровно `(k - 1)` человек
- На одном этапе: вероятность получить в пару слабого соперника = `(k - 1) / 79`

**Почему делю на 79?** Потому что соперник выбирается случайно из 79 возможных.

**Два этапа:**
- Этапы независимые (пары каждый раз формируются заново)
- Вероятность выиграть оба этапа = (вероятность выиграть 1-й) × (вероятность выиграть 2-й) = `((k - 1) / 79)²`

*Самый сильный повар (ранг 80) может выиграть с вероятностью (79/79)² = 1 — он всегда побеждает всех. А самый слабый (ранг 1) выигрывает с вероятностью 0 — он никого не может победить. Остальные — где-то посередине. В среднем получается около 27 победителей. Это логично: сильных поваров больше, чем слабых, поэтому победителей достаточно много.*

### Задание 3: Одинокая дорога

**Условие:**
На пустынном шоссе вероятность появления автомобиля за 30-минутный период составляет 0.95. Какова вероятность его появления за 10 минут? А за 27 минут? Ответ дать в процентах, округлив до десятых через точку с запятой, например: 42,7; 95,0.

````python
# Базовая вероятность отсутствия машин за 30 минут
q_30 = 1 - 0.95

# Расчет вероятностей для 10 и 27 минут (в процентах)
p_10 = (1 - q_30 ** (10 / 30)) * 100
p_27 = (1 - q_30 ** (27 / 30)) * 100

# Форматированный вывод по ТЗ
print(f"Ответ: {p_10:.1f}; {p_27:.1f}".replace('.', ','))
````

**Логика решения:**

1. **Переворачиваем задачу** — легче считать не "машина появилась", а "машина НЕ появилась".
   - За 30 минут: P(нет машин) = 1 − 0,95 = 0,05

2. **Делаем разумное предположение** — машины появляются независимо друг от друга (как в пуассоновском потоке). Это значит:
   - Если за 1 минуту P(нет машин) = q
   - То за 30 минут: q³⁰ = 0,05
   - Значит, q = 0,05^(1/30)

3. **Для любого времени t:**
   - P(нет машин за t минут) = q^t = 0,05^(t/30)
   - P(есть машина) = 1 − 0,05^(t/30)

4. **Считаем:**
   - За 10 минут: 1 − 0,05^(10/30) = 1 − 0,05^(1/3) ≈ 0,632 = 63,2%
   - За 27 минут: 1 − 0,05^(27/30) = 1 − 0,05^(0,9) ≈ 0,933 = 93,3%
     
*"Я рассуждала так: проще считать вероятность, что машины НЕТ. За 30 минут она равна 0,05. Если предположить, что появление машин независимо во времени, то за 1 минуту эта вероятность — корень 30-й степени из 0,05. А за t минут — 0,05 в степени t/30. Дальше подставляю 10 и 27 минут, вычитаю из 1 и получаю 63,2% и 93,3%."*

### Задание 1: Изоморфизмы

**Условие:**
Реализовать функцию (или тело функции), которая проверяет на изоморфность два слова. 

**Пояснение:** строки `s` и `t` называются изоморфными, если все вхождения каждого символа строки `s` можно последовательно заменить другим символом и получить строку `t`. Порядок символов при этом должен сохраняться, а замена — быть уникальной. Так, два разных символа строки `s` нельзя заменить одним и тем же символом из строки `t`, а вот одинаковые символы в строке `s` должны заменяться одним и тем же символом.

**Пример:**
```python
s = 'paper'
t = 'title'
print(is_isomorphic(s, t))
```

**Вывод:**
```text
True
```

*Оценить оптимальность решения по времени и памяти и прикрепить текст кода.*

**Как решу:**

Нужно проверить два условия:
1. Если символ из `s` уже встречался — он должен заменяться на тот же символ из `t`.
2. Если символ из `t` уже встречался — к нему должен подходить тот же символ из `s`.

**Почему два условия?** Потому что замена должна быть взаимно-однозначной. Например, `"ab"` → `"aa"`: `a→a, b→a` — два разных символа из `s` заменяются на один символ из `t`. Это запрещено. Второй словарь `t_to_s` ловит эту ошибку.

**Алгоритм:**
1. Проверяем длины — если разные, сразу `False`.
2. Идём по парам символов `(char_s, char_t)`.
3. Смотрим в словарь `s_to_t`:
   - Если `char_s` уже есть — проверяем, что он соответствует текущему `char_t`.
   - Если нет — запоминаем замену.
4. Смотрим в словарь `t_to_s`:
   - Если `char_t` уже есть — проверяем, что он соответствует текущему `char_s`.
   - Если нет — запоминаем.
5. Если дошли до конца без ошибок — `True`.

````python
def is_isomorphic(s: str, t: str) -> bool:
    # Если длины строк разные, они не могут быть изоморфными
    if len(s) != len(t):
        return False
        
    s_to_t = {}
    t_to_s = {}
    
    for char_s, char_t in zip(s, t):
        # Проверяем отображение из s в t
        if char_s in s_to_t:
            if s_to_t[char_s] != char_t:
                return False
        else:
            s_to_t[char_s] = char_t
            
        # Проверяем отображение из t в s
        if char_t in t_to_s:
            if t_to_s[char_t] != char_s:
                return False
        else:
            t_to_s[char_t] = char_s
            
    return True

# Пример использования:
s = 'paper'
t = 'title'
print(is_isomorphic(s, t))
````

**Временная сложность (Time Complexity):** $O(N)$, где $N$ — длина строк. Алгоритм выполняет один проход по строкам с помощью `zip()`. Все операции поиска и вставки в словари выполняются в среднем за $O(1)$, поэтому общая сложность составляет $O(N)$.

**Пространственная сложность (Space Complexity):** $O(\min(N, K))$, где $N$ — длина строки, а $K$ — количество различных символов, которые могут встретиться во входных данных. Дополнительная память используется для двух словарей (`s_to_t` и `t_to_s`), в которых хранятся соответствия между символами. В худшем случае количество сохранённых пар не превышает числа уникальных символов во входных строках.

*Алгоритм проходит по строкам один раз и поддерживает двустороннее отображение между символами с помощью двух словарей. Это позволяет проверить взаимно-однозначное соответствие символов за линейное время и избежать ситуаций, когда разные символы одной строки отображаются в один и тот же символ другой строки.*

### Задание 2: Натуральная последовательность

**Условие:** Реализовать функцию (или тело функции), которая находит единственное отсутствующее число из последовательности натуральных чисел \(1, 2, \dots, n\).

**Пример:**
```python
nums = [1, 2, 3, 4, 5, 6, 8, 9, 10, 11]
print(missing_number(nums))
```

**Вывод:**
```text
7
```

*Оценить оптимальность решения по времени и памяти и прикрепить текст кода.*


````python
def missing_number(nums: list[int]) -> int:
    # Длина массива на 1 меньше, чем n, так как одно число пропущено
    n = len(nums) + 1
    
    # Сумма идеальной последовательности от 1 до n по формуле Гаусса
    expected_sum = n * (n + 1) // 2
    
    # Фактическая сумма элементов в массиве
    actual_sum = sum(nums)
    
    # Разница и есть пропущенное число
    return expected_sum - actual_sum

# Пример из условия:
nums = [1, 2, 3, 4, 5, 6, 8, 9, 10, 11]
print(missing_number(nums))
````

**Временная сложность (Time Complexity):** $O(N)$, где $N$ — количество элементов массива. Основная операция — вычисление суммы элементов с помощью `sum(nums)`, которое требует одного прохода по массиву. Остальные вычисления выполняются за константное время $O(1)$.

**Пространственная сложность (Space Complexity):** $O(1)$. Алгоритм использует фиксированное количество дополнительных переменных (`n`, `expected_sum`, `actual_sum`) и не создаёт структур данных, размер которых зависит от объёма входных данных.

*Алгоритм использует формулу суммы первых $n$ натуральных чисел. Сначала вычисляется ожидаемая сумма последовательности от 1 до $n$, затем из неё вычитается фактическая сумма элементов массива. Полученная разность и является пропущенным числом.*

### Задание 3: Факторизация

**Условие:** Реализовать функцию (или тело функции), которая при введении натурального числа `n` разбивает его на простые множители (представить его в виде простых чисел).

**Пример:**
```python
n = 56
print(prime_factors(n))
```

**Вывод:**
```text
[2, 2, 2, 7]
```

*Оценить оптимальность решения по времени и памяти и прикрепить текст кода.*

````python
def prime_factors(n: int) -> list[int]:
    factors = []
    # Начинаем проверку с первого простого числа
    d = 2
    # Перебираем делители до тех пор, пока квадрат делителя не превысит n
    while d * d <= n:
        while n % d == 0:
            factors.append(d)
            n //= d
        d += 1
    # Если после цикла n больше 1, значит само n — простое число
    if n > 1:
        factors.append(n)
    return factors

# Пример из условия:
n = 56
print(prime_factors(n))
````

### Оценка сложности

**Временная сложность (Time Complexity):** $O(\sqrt{n})$.

Алгоритм последовательно проверяет возможные делители числа, начиная с 2. В худшем случае (когда число является простым) потребуется проверить все делители до $\sqrt{n}$, поскольку если у числа есть делитель больше $\sqrt{n}$, то соответствующий ему парный делитель обязательно меньше $\sqrt{n}$.

**Пространственная сложность (Space Complexity):** $O(k)$,

где $k$ — количество найденных простых множителей. Дополнительная память используется только для хранения результирующего списка множителей. Помимо этого алгоритм использует фиксированное количество вспомогательных переменных.

### Задание 3: Покупки

**Условие:**
```sql
create table account
(
id integer, -- ID счета
client_id integer, -- ID клиента
open_dt date, -- дата открытия счета
close_dt date -- дата закрытия счета
)

create table transaction
(
id integer, -- ID транзакции
account_id integer, -- ID счета
transaction_date date, -- дата транзакции
amount numeric(10,2), -- сумма транзакции
type varchar(3) -- тип транзакции
)
```
Требуется вывести ID клиентов, которые за последний месяц по всем своим счетам совершили покупок меньше, чем на 5000 рублей. Без использования подзапросов и оконных функций.


````python
SELECT
    a.client_id
FROM account a
JOIN transaction t
    ON a.id = t.account_id
    AND t.transaction_date >= CURRENT_DATE - INTERVAL '1 month'
    AND t.transaction_date BETWEEN a.open_dt AND COALESCE(a.close_dt, CURRENT_DATE)
    AND t.type = 'BUY'
GROUP BY
    a.client_id
HAVING
    COALESCE(SUM(t.amount), 0) < 5000;
````

**Логика решения:**

1. **Соединяем счета и транзакции** — чтобы определить, какие покупки совершал каждый клиент по своим счетам.

2. **Отбираем только покупки за последний месяц** — с помощью условия `transaction_date >= CURRENT_DATE - INTERVAL '1 month'`.

3. **Учитываем период существования счёта** — транзакция должна быть совершена после открытия счёта и до его закрытия. Для этого проверяем, что дата транзакции находится между `open_dt` и `close_dt`. Если счёт ещё не закрыт (`close_dt IS NULL`), используем `COALESCE(close_dt, CURRENT_DATE)`.

4. **Группируем данные по клиентам** — чтобы посчитать общую сумму покупок по всем счетам каждого клиента.

5. **Оставляем только клиентов с суммой покупок менее 5000 рублей** — с помощью условия в `HAVING`.

6. **Используем агрегатную функцию SUM()** для расчёта общей суммы покупок клиента. В текущем запросе `COALESCE(SUM(t.amount), 0)` не влияет на результат, поскольку используется `INNER JOIN`, и клиенты без покупок не попадают в выборку. Если требуется учитывать таких клиентов и считать их сумму покупок равной нулю, следует использовать `LEFT JOIN`.

---

**Важные уточнения:**

- Предполагаю, что покупки определяются значением `type = 'BUY'`.
- Запрос считает сумму покупок по всем счетам клиента, поскольку группировка выполняется по `client_id`.
- В текущем варианте используются только клиенты, у которых есть хотя бы одна покупка за рассматриваемый период. Для включения клиентов без покупок потребуется `LEFT JOIN`.

**Условие:** Что такое p-value в статистическом тестировании? Выберите одно верное утверждение:
1. Вероятность, что нулевая гипотеза верна.
2. Вероятность наблюдения такого или более экстремального результата при условии, что нулевая гипотеза верна.
3. Значение уровня значимости, при котором отвергается нулевая гипотеза.

**Выбранное верное утверждение:** 
* **2. Вероятность наблюдения такого или более экстремального результата при условии, что нулевая гипотеза верна.**

**Обоснование:**
* **Почему это верно:** Это классическое математическое определение $p$-value. Оно показывает, какова вероятность получить наши тестовые данные (или еще более сильные отклонения) исключительно за счет случайности, если на самом деле никакого реального эффекта нет (то есть справедлива нулевая гипотеза $H_0$).
* **Почему не верны остальные варианты:**
  * *Вариант 1* — частое заблуждение. $p$-value не оценивает вероятность истинности самой гипотезы, оно оценивает вероятность *данных* при условии гипотезы: $P(\text{Данные} \mid H_0)$, а не $P(H_0 \mid \text{Данные})$.
  * *Вариант 3* — уровень значимости ($\alpha$) задается исследователем заранее (например, 0.05) и является фиксированным порогом, с которым $p$-value лишь сравнивается в конце.
  * *Вариант 4* — среднее значение выборки является точечной метрикой ($\bar{X}$) и никак не связано с вероятностным показателем теста.

4. Большой размер выборок позволяет использовать параметрические методы благодаря центральной предельной теореме (ЦПТ), поэтому проблема заключается не в ненормальности исходного распределения, а в возможном различии дисперсий.
5. Непараметрические критерии (например, Манна–Уитни) также могут использоваться, но при таком объёме данных и интересе к сравнению средних тест Уэлча является более естественным выбором.

### Задание 2: Ручной счёт ROC_AUC

**Условие:** Классификатор выдал следующие прогнозируемые метки класса и вероятности принадлежности к классу "1". На основе полученных данных рассчитайте метрику ROC_AUC. Тезисно описать ход решения.

*Ответ округлить до сотых, например: 4,12*


| Истинная метка класса | Порог классификации (0.6) | Оценка вероятности |
| :--- | :--- | :--- |
| 1 | 1 | 0.95 |
| 0 | 1 | 0.90 |
| 1 | 1 | 0.85 |
| 0 | 1 | 0.80 |
| 1 | 1 | 0.75 |
| 1 | 1 | 0.70 |
| 1 | 1 | 0.65 |
| 1 | 1 | 0.60 |
| 0 | 0 | 0.55 |
| 0 | 0 | 0.50 |
| 0 | 0 | 0.45 |
| 1 | 0 | 0.40 |
| 0 | 0 | 0.35 |
| 0 | 0 | 0.30 |
| 0 | 0 | 0.25 |


````python
y_true = [1, 0, 1, 0, 1, 1, 1, 1, 0, 0, 0, 1, 0, 0, 0]
y_score = [0.95, 0.90, 0.85, 0.80, 0.75, 0.70, 0.65, 0.60, 0.55, 0.50, 0.45, 0.40, 0.35, 0.30, 0.25]

# отделяем positive и negative
pos = [s for s, y in zip(y_score, y_true) if y == 1]
neg = [s for s, y in zip(y_score, y_true) if y == 0]

correct_pairs = 0
total_pairs = len(pos) * len(neg)

# считаем попарные сравнения
for p in pos:
    for n in neg:
        if p > n:
            correct_pairs += 1
        elif p == n:
            correct_pairs += 0.5

roc_auc = correct_pairs / total_pairs

print(f"ROC_AUC: {roc_auc:.2f}")
````

**Ход решения:**
ROC-AUC интерпретируется как вероятность того, что случайный объект класса 1 имеет более высокий скор, чем случайный объект класса 0.

Всего положительных объектов: 7  
Всего отрицательных объектов: 8  
Всего пар: 56  

Для каждого объекта класса 1 считаем количество объектов класса 0 с меньшим скором:

0.95 - 8  
0.85 - 7  
0.75 - 6  
0.70 - 6  
0.65 - 6  
0.60 - 6  
0.40 - 3  

Сумма правильных пар = 42  

ROC_AUC = 42 / 56 = 0.75

*Колонка "Порог классификации (0.6)" не используется при расчёте ROC-AUC.*

*ROC-AUC оценивает качество ранжирования модели на основе предсказанных вероятностей и истинных меток классов, без привязки к фиксированному порогу классификации.*

**Поэтому в расчёте используются только истинные метки и значения вероятностей.**

*Колонка с порогом отражает лишь результат бинаризации при threshold = 0.6 и не влияет на значение ROC-AUC, так как метрика учитывает все возможные пороги одновременно.*

### Задание 3: Ручной счёт корреляции

**Условие:** Рассчитайте линейную корреляцию Пирсона на основе предоставленных данных. Какой вывод можно сделать на основе полученного результата? Можно ли утверждать, что существует причинно-следственная связь между количеством чашек кофе, выпитых студентами в течение экзаменационного дня, и их итоговым баллом за экзамен?

*Ответ округлить до сотых, например: 4,12*


| Число выпитых чашек кофе | Балл за экзамен |
| :--- | :--- |
| 1 | 85 |
| 1 | 88 |
| 2 | 79 |
| 2 | 81 |
| 2 | 84 |
| 2 | 65 |
| 3 | 67 |
| 3 | 58 |
| 3 | 76 |
| 4 | 49 |


````python
import numpy as np

# Загружаем данные из таблицы
x = np.array([1, 1, 2, 2, 2, 2, 3, 3, 3, 4])
y = np.array([85, 88, 79, 81, 84, 65, 67, 58, 76, 49])

# Рассчитываем коэффициент корреляции Пирсона
pearson_corr = np.corrcoef(x, y)[0, 1]

print(f"Коэффициент корреляции Пирсона: {pearson_corr:.2f}")
````

**Вывод на основе результата:** r = -0.85

Полученное значение коэффициента корреляции Пирсона указывает на сильную отрицательную линейную зависимость между количеством выпитого кофе и итоговым баллом за экзамен.

Однако корреляция не позволяет делать вывод о причинно-следственной связи.

На основании этих данных нельзя утверждать, что увеличение потребления кофе приводит к снижению экзаменационного балла. Наблюдаемая зависимость может объясняться наличием скрытых факторов (confounders), таких как уровень подготовки студентов, стресс или недостаток сна, которые одновременно влияют и на потребление кофе, и на результат экзамена.
